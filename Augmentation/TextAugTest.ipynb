```python
from ktextaug.transformative.noise_addition import vowel_noise, phonological_process  # function
import pandas as pd
```

```python
def textProcessingVowelNoise(sentence, subject_entity, object_entity):
    sentence = sentence.replace(subject_entity, '#')
    sentence = sentence.replace(object_entity, '@')
    sentence = vowel_noise(sentence, prob=1)
    sentence = sentence.replace('#', subject_entity)
    sentence = sentence.replace('@', object_entity)
    return sentence

def textProcessingPhonologicalProcess(sentence, subject_entity, object_entity):
    sentence = sentence.replace(subject_entity, '#')
    sentence = sentence.replace(object_entity, '@')
    sentence = phonological_process(sentence, prob=1)
    sentence = sentence.replace('#', subject_entity)
    sentence = sentence.replace('@', object_entity)
    return sentence
```

```python
dataset_dir = '../dataset/train/train_finalCorrection.csv'
original_dataset = pd.read_csv(dataset_dir)
```

```python
original_dataset
```

```python
sentences = []

for i,j,k in zip(original_dataset['subject_entity'], original_dataset['object_entity'], original_dataset['sentence']):
    dict_i = eval(i) # str을 코드화
    dict_j = eval(j)
    i = dict_i['word'] # subj
    j = dict_j['word'] # obj
    
    k = textProcessingVowelNoise(k, i, j)

    sentences.append(k)
```

```python
out_dataset = pd.DataFrame({'id':original_dataset['id'], 'sentence':sentences,'subject_entity':original_dataset['subject_entity'],'object_entity':original_dataset['object_entity'],'label':original_dataset['label'],'source':original_dataset['source']})
```

```python
out_dataset.to_csv('augmented_vowelNoise.csv', index=False)
```
