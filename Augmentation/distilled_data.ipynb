```python
original_dataset.drop_duplicates(['sentence', 'subject_entity', 'object_entity'])
```

```python
subj_word = []
obj_word = []

for i in range(original_dataset.shape[0]):
    subj_word.append(eval(original_dataset['subject_entity'].iloc[i])["word"])
    obj_word.append(eval(original_dataset['object_entity'].iloc[i])["word"])
```

```python
modified_dataset = original_dataset

modified_dataset['subj_word'] = subj_word
modified_dataset['obj_word'] = obj_word
modified_dataset['subject_entity'] = original_dataset['subject_entity']
modified_dataset['object_entity'] = original_dataset['object_entity']
```

```python
temp = modified_dataset.drop_duplicates(['sentence', 'subj_word', 'obj_word'])
```

```python
ttemp = temp.drop(columns=['subj_word', 'obj_word'])
ttemp
```

```python
ttemp.to_csv('train_removed_duplicated_sent_sub_obj.csv', index=False)
```
